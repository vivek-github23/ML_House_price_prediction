```python
import pandas as pd
import re

def count_urls(text):
    url_pattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    urls = re.findall(url_pattern, text)
    return len(urls)

# Assuming 'review_text' is the column containing text in your DataFrame
df['url_count'] = df['review_text'].apply(lambda x: count_urls(x))

# Display the DataFrame with the added 'url_count' column
count = 0
for i in range(df['url_count'].size):
    count += df['url_count'][i]

print(count)
```

```python
def remove_urls(text):
    url_pattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    return re.sub(url_pattern, '', text)

# Apply the function to the 'review_text' column
df['review_text'] = df['review_text'].apply(remove_urls)
```

Expanding contradicting words like **didn't to did not**

```python
import re
contradiction_mapping = {
    "didn't": "did not",
    "won't": "will not",
    "can't": "can not",
    "doesn't": "does not",
    "don't": "do not",
    "didn": "did not",
    "don'": "do not",
    "doesn'": "does not",
    "aren't": "are not",
    "isn't": "is not",
    "wasn't": "was not",
    "weren't": "were not",
    "aren'": "are not",
    "haven't": "have not",
    "hasn't": "has not",
    "hadn't": "had not",
    "haven'": "have not",
    "hasn'": "has not",
    "hadn'": "had not",
    "won't": "will not",
    "wouldn't": "would not",
    "can't": "can not",
    "cannot": "can not",
    "couldn't": "could not",
    "weren'": "were not",
    "wouldn'": "would not",
    "shan't": "shall not",
    "shouldn't": "should not",
    "mustn't": "must not",
    "shouldn'": "should not",
    "shan'": "shall not",
    "mustn'": "must not",
    "mightn'": "might not",
    "isn'": "is not"
}
def replace_contractions(text, mapping):
    for contraction, replacement in mapping.items():
        # Use re.IGNORECASE for case-insensitive replacement
        text = re.sub(r'\b' + re.escape(contraction) + r'\b', replacement, text, flags=re.IGNORECASE)
    return text

df['review_text'] = df['review_text'].apply(lambda x: replace_contractions(x, contradiction_mapping))

```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download('wordnet')

def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    words = word_tokenize(text)
    lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in words]  # Specify POS='v' for verbs
    return ' '.join(lemmatized_words)
```

```python
df['review_text'] = df['review_text'].apply(lambda x: lemmatize_text(x))

df['review_text'].head(20)
```

```python


def preprocess_text(text):
    
    #translator = str.maketrans('', '', string.punctuation)
    # Tokenize the text
    words = word_tokenize(text)
    
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Join the filtered words back into a single string
    return ' '.join(filtered_words)
```

```python
df['review_text'] = df['review_text'].apply(preprocess_text)
```
